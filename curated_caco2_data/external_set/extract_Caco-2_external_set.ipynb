```python
from rdkit import Chem,DataStructs
from rdkit.Chem import MolStandardize
from rdkit.Chem import AllChem
import numpy as np
import pandas as pd
from math import log10
```

```python
df_22_23 = pd.read_csv('./chembl_Papp_2022_2023_raw.txt',sep='\t')
```

```python
df_22_23 = df_22_23[df_22_23['Standard_Value']!=0]
```

```python
df_22_23
```

```python
df_22_23['logPapp'] = df_22_23['Standard_Value'].apply(lambda x:log10(x*1e-6))
```

```python
df_22_23 = df_22_23[['Smiles', 'logPapp']] 
```

```python
df_22_23
```

```python
smilesList=df_22_23.Smiles.values
print("number of all smiles:",len(smilesList))

canonical_smiles_list=[]
for smiles in smilesList:
    try:
        mol=Chem.MolFromSmiles(smiles)
        mol=MolStandardize.rdMolStandardize.Cleanup(mol)
        mol=MolStandardize.rdMolStandardize.FragmentParent(mol)

        canonical_smiles_list.append(Chem.MolToSmiles(mol,isomericSmiles=True))
    except:
        print(smiles)
        pass            
```

```python
df_22_23['cano_smi'] = canonical_smiles_list
```

```python
df_22_23
```

```python
duplicates = df_22_23.cano_smi.value_counts()[df_22_23.cano_smi.value_counts()>=2]
```

```python
uniq = df_22_23.cano_smi.value_counts()[df_22_23.cano_smi.value_counts()< 2]
```

```python
len(duplicates),len(uniq)
```

```python
from collections import defaultdict

duplicates_dict=defaultdict(dict)

for i,dup in enumerate(duplicates.index):
    duplicates_dict[i]['smiles'] = dup
    duplicates_dict[i]['avg'] = np.mean(df_22_23[df_22_23.cano_smi == dup]['logPapp'].values)
    duplicates_dict[i]['std'] = np.std(df_22_23[df_22_23.cano_smi == dup]['logPapp'].values)
    duplicates_dict[i]['count'] = len(df_22_23[df_22_23.cano_smi == dup]['logPapp'].values)
```

```python
duplicates_df = pd.DataFrame(duplicates_dict).T
```

```python
duplicates_df
```

```python
df_0_03 = duplicates_df[(duplicates_df['std'] > 0) & (duplicates_df['std'] < 0.3)] 
```

```python
df_0 = duplicates_df[duplicates_df['std'] == 0]
```

```python
df_0
```

```python
df_0 = df_0[['smiles', 'avg']]
df_0_03 = df_0_03[['smiles', 'avg']]
```

```python
from collections import defaultdict

uniq_dict=defaultdict(dict)

for i,dup in enumerate(uniq.index):
     uniq_dict[i]['smiles'] = dup
     uniq_dict[i]['avg'] = df_22_23[df_22_23.cano_smi == dup]['logPapp'].values[0]
```

```python
uniq_df = pd.DataFrame(uniq_dict).T
```

```python
uniq_df
```

```python
external_set = pd.concat([uniq_df, df_0,df_0_03]).rename(columns={'avg': 'logPapp'})
```

```python
external_set['logPapp'].min(),external_set['logPapp'].max(),external_set['logPapp'].mean()
```
